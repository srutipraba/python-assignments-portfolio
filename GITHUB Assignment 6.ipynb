```python
#All my imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
def normal_density1 (mean,variance,x):
    function = (1/(np.sqrt((2*np.pi)*variance)))*((np.e)**(-((x-mean)**2)/(2*variance)))
    return function
#running the normal density function
mean=2
variance=3
x_values=np.linspace(-10,10,100)
y_values = normal_density1(mean, variance, x_values)
#calculating x and y values given the mean and variance
plt.plot(x_values, y_values, 'bo')
plt.title('Normal Density Function')
plt.xlabel('x')
plt.ylabel('Density')
plt.grid(True)
plt.show()
#plotting the function
```

```python
#Please note that I ran the integration function 3 different ways (left Reimann sums, right and midpoint) just for accuracy
def normal_density1 (mean,variance,x):
    function = (1/(np.sqrt((2*np.pi)*variance)))*((np.e)**(-((x-mean)**2)/(2*variance)))
    return function
def integration5(mean, variance, lower_limit, upper_limit):
    n = 50
    x = np.linspace(lower_limit,upper_limit,n+1) 
    y_lefthandside = []  
    y_righthandside = []
    y_midpoint =[] 
    delta_x = (upper_limit-lower_limit)/n
    #defining my range, x_values, delta_x, and lists to contain my y values from all reimann sums
    # left sum
    for i in range(n):
        xl = x[i]
        yl = (1/(np.sqrt((2*np.pi)*(variance))))*((np.e)**(-((xl-mean)**2)/(2*variance)))
        y_lefthandside.append(yl)
    ls = sum(y_lefthandside)*delta_x
    print("The left sum is:", ls)

    # right sum
    for i in range(n):
        xr = x[i+1]
        yr = (1/(np.sqrt((2*np.pi)*(variance))))*((np.e)**(-((xr-mean)**2)/(2*variance)))
        y_righthandside.append(yr)
    rs = sum(y_righthandside)*delta_x
    print("The right sum is:", rs)

    # midpoint sum
    for i in range(n):
        xm = ((x[i])+(x[i+1]))/2
        ym = (1/(np.sqrt((2*np.pi)*(variance))))*((np.e)**(-((xm-mean)**2)/(2*variance)))
        y_midpoint.append(ym)
    ms = sum(y_midpoint)*delta_x

    print("The midpoint sum is:", ms)
```
