```python
def exercise_function_machine(x,y):
    exercise_function=(np.e**x)*(sp.sin(y))+y**3
    derivative_of_exercise_function_x=sp.diff(exercise_function,x) #partial derivative of the function with respect to x
    derivative_of_exercise_function_y=sp.diff(exercise_function,y) #partial derivative of the function with respect to y
    return derivative_of_exercise_function_x and derivative_of_exercise_function_y
#my comments are integrated within my assignments, but this computes the partial derivatives of f with respect to x and then y and returns it
```

```python
exercise_function_2=(x**2)*y+x*y**2
derivative_of_exercise_function_2_x=sp.diff(exercise_function_2,x) #partial derivative of the function with respect to x
derivative_of_exercise_function_2_y=sp.diff(exercise_function_2,y) #partial derivative of the function with respect to y
print("This is the derivative of exercise_2:",(derivative_of_exercise_function_2_x,derivative_of_exercise_function_2_y))
actual_derivative_of_exercise_function_2_x = derivative_of_exercise_function_2_x.subs([(x,1),(y,-1)]) #evaluates the partial derivative of f with respect to x at a specific point
actual_derivative_of_exercise_function_2_y = derivative_of_exercise_function_2_y.subs([(x,1),(y,-1)]) #evaluates the partial derivative of f with respect to y at a specific point
print("This is the evaluated derivative of exercise_2:",(actual_derivative_of_exercise_function_2_x, actual_derivative_of_exercise_function_2_y))
magnitude_of_actual_derivative_of_exercise_function_2=np.sqrt((int(actual_derivative_of_exercise_function_2_x))**2+(int(actual_derivative_of_exercise_function_2_y))**2) #calculates the magnitude of both partial derivatives
print("This is the magnitude of the derivative of exercise_2:",magnitude_of_actual_derivative_of_exercise_function_2)
#this code section provides the partial derivatives of f with respect to x and y, evaluates the derivative of the functions and computes the magnitude of it (using our typical distance formula)
#this is for the function (x**2)*y+x*y**2
```

```python
x=sp.symbols('x')
y=sp.symbols('y')
exercise_function_3=sp.log(x**2+y**2)
derivative_of_exercise_function_3_x=sp.diff(exercise_function_3,x) #partial derivative of the function with respect to x
derivative_of_exercise_function_3_y=sp.diff(exercise_function_3,y)#partial derivative of the function with respect to y
derivative_of_exercise_function_3_x_x=sp.diff(derivative_of_exercise_function_3_x,x) #fxx computed here
derivative_of_exercise_function_3_y_y=sp.diff(derivative_of_exercise_function_3_y,y) #fyy computed here
derivative_of_exercise_function_3_x_y=sp.diff(derivative_of_exercise_function_3_x,y) #fxy computed here
derivative_of_exercise_function_3_y_x=sp.diff(derivative_of_exercise_function_3_y,x) #fyx  computed here
print("These are the second derivatives of exercise_function_3")
print("1) df/dx^2=",derivative_of_exercise_function_3_x_x) #printing all 4 partial derivatives
print("2) df/dy^2=",derivative_of_exercise_function_3_y_y)
print("3) df/dxy=",derivative_of_exercise_function_3_x_y)
print("4) df/dyx=",derivative_of_exercise_function_3_y_x)
#this code computes the first partial derivatives of log(x**2+y**2) with respect to x and then y, and then the second partial derivatives and prints them out.
```

```python
x, y = sp.symbols('x y')
j = x**3 - 3*x*y + y**3
j_func = sp.lambdify((x, y), j, 'numpy')
x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)
plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
#this code just provides a contour plot of x**3 - 3*x*y + y**3
```
