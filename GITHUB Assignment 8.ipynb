```python
import numpy as np
def total_gradient_descent2(x_initial,y_initial, learning_rate, number_of_iterations):
    x=x_initial
    y=y_initial
    gradients=[]
    for c in range(number_of_iterations):
        gradient_x = 2*x*np.e**(-x**2-(y-2)**2+4*x*np.e**(-x**2-(y+2)**2))
        gradient_y = (np.exp(-x**2-(y-2)**2))*2*(y-2)+4*(y+2)*np.exp(-x**2-(y+2)**2)
        gradients.append((x,y))
        x=x-learning_rate*gradient_x
        y=y-learning_rate*gradient_y
    final_gradient=(x,y)
    return final_gradient
#same gradient function - just has the next function on the assignment (1-np.exp(-x**2-(y-2)**2)-2*np.exp(-x**2-(y+2)**2))
def function_2(x,y):
    return 1-np.exp(-x**2-(y-2)**2)-2*np.exp(-x**2-(y+2)**2)
#this just defines the function again
import numpy as np
from mpl_toolkits import mplot3d #for 3D plots
import matplotlib.pyplot as plt #usual matplotlib
%matplotlib widget
X=np.linspace(-5,5,100)
Y=np.linspace(-5,5,100)
x,y=np.meshgrid(X,Y)
z= 1-np.exp(-x**2-(y-2)**2)-2*np.exp(-x**2-(y+2)**2)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z,cmap='viridis', edgecolor='none')
#x,y z are variable names.
```
